# Test Stock Query API
Ejemplo de cómo invocar los endpoints de auth y query.

```python
import httpx

BASE_URL = "http://localhost:8000"
```

## 1. Health check

```python
async with httpx.AsyncClient() as client:
    r = await client.get(f"{BASE_URL}/health")
    print(r.json())
```

## 2. Get login URL

```python
async with httpx.AsyncClient() as client:
    r = await client.get(f"{BASE_URL}/auth/login")
    print(r.json())
```

## 3. Authenticate (get tokens)

```python
USERNAME = "your-username"
PASSWORD = "your-password"

async with httpx.AsyncClient() as client:
    r = await client.post(
        f"{BASE_URL}/auth/token",
        json={"username": USERNAME, "password": PASSWORD},
    )
    tokens = r.json()
    print(tokens)
    access_token = tokens["access_token"]
```

## 4. Get user info

```python
headers = {"Authorization": f"Bearer {access_token}"}

async with httpx.AsyncClient() as client:
    r = await client.get(f"{BASE_URL}/auth/user", headers=headers)
    print(r.json())
```

## 5. Query stock agent

```python
async with httpx.AsyncClient() as client:
    r = await client.post(
        f"{BASE_URL}/query",
        headers=headers,
        json={"question": "What is Amazon's current stock price?"},
    )
    print(r.json())
```
